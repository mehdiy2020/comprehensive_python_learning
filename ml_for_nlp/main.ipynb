## Basic Terminologies


### Corpus
A Paragraph

### Documents
Sentences

### Vocabulary
Unique Words Presented in dictionary

### Words
All the Words in Corpus

## Tokenization

- When we take either a paragraph (corpus) pr sentences (Documents) and convert them into TOKENS.
- Next Level of TOkenization is to convert each documents (sentenses) into Words
- Total non-repeated words called VOcabularies


```python
# Tokenization

corpus="""
A general order of operator precedence in Python, from highest to lowest:
1. Parentheses `()`
2. Exponentiation `**`
3. Unary operators `+`, `-` (positive, negative)
4. Multiplication, Division, Floor Division, and Modulus `*`, `/`, `//`, `%`
5. Addition and Subtraction `+`, `-`
6. Relational Operators `==`, `!=`, `<`, `<=`, `>`, `>=`
7. Membership Operators `in`, `not in`
8. Identity Operators `is`, `is not`
9. Logical NOT `not`
10. Logical AND `and`
11. Logical OR `or
Expressions are combinations of values, variables, and operators that, when evaluated, produce a result. Think of them as mathematical equations that the computer solves to give us an answer.
Operators are special symbols in Python that carry out arithmetic, logical, or relational computations. They act as a bridge, connecting values and variables to produce a desired result.

"""
print(corpus)
import nltk
nltk.download('punkt') # Pre-trained data model
nltk.download('punkt_tab')

```

```python
## Convert Paragraphs into Sentences


from nltk.tokenize import sent_tokenize 

documents = sent_tokenize(corpus)
type(documents)

```

```python
for snetence in documents:
  print(snetence)
```

```python
## Convert Paragraphs --> Words
## Convert Sentence --> Words

from nltk.tokenize import word_tokenize

word_tokenize(corpus)
```

```python
for snetence in documents:
  print(word_tokenize(snetence))
```

```python
from nltk.tokenize import wordpunct_tokenize

wordpunct_tokenize(corpus)
```

```python
for snetence in documents:
  print(wordpunct_tokenize(snetence))
```

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

tokenizer.tokenize(corpus)
```

1. spaCy: The Production & Speed Library 🚀
If NLTK is a university research lab (great for experimenting), spaCy is an industrial factory (great for building fast, efficient products). It's designed to be fast, modern, and ready for real-world applications.


Best For: Production systems, web applications, and any task where speed and efficiency are critical.

Key Features:

Blazing fast tokenization, parsing, and Named Entity Recognition (NER).

Comes with pre-trained statistical models for many languages.

Object-oriented design (you work with Doc and Token objects), which is often more intuitive than NLTK's string-based approach.

Excellent for building data pipelines that process a lot of text.

Use spaCy when: You are building an application and need reliable, high-performance NLP.

2. Hugging Face transformers: The State-of-the-Art 🧠
This is arguably the most important library in modern NLP. It gives you easy access to thousands of powerful, pre-trained Transformer models like BERT, GPT-2, and T5. These models are the basis for almost all cutting-edge NLP tasks today.

Best For: Any task requiring deep language understanding, like sentiment analysis, question answering, text generation, and summarization.

Key Features:

Model Hub: A huge repository of pre-trained models you can download and use in just a few lines of code.

Pipelines: A simple API for complex tasks. For example: pipeline("sentiment-analysis")

Interoperable with both PyTorch and TensorFlow.

Use transformers when: You need the highest accuracy and want to use modern, state-of-the-art models.
